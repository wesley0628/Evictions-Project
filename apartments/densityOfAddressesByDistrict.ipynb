```python
totalExcludeSameAddressList = list()
totalAddressList = list()

for i in range(1,21):
    totalExcludeSameAddressList.append(len(excludeSameAddressDF[excludeSameAddressDF['Aldermanic District'] == i]))
    totalAddressList.append(len(includeSameAddressDF[includeSameAddressDF['Aldermanic District'] == i]) + len(excludeSameAddressDF[excludeSameAddressDF['Aldermanic District'] == i]))
```

```python
totalExcludeSameAddressList
```

```python
totalAddressList
```

```python
densityOfExcludeSameAddressList = list()
for i, j in zip(totalExcludeSameAddressList, totalAddressList):
    densityOfExcludeSameAddressList.append(round(i/j*100,2))
```

```python
densityOfExcludeSameAddressList
```

```python
densityOfExcludeSameAddressSeries = pd.Series(densityOfExcludeSameAddressList)
```

```python
import matplotlib.patches as mpatches
import mapclassify as mc
import matplotlib.pyplot as plt
import numpy as np

# To create label on each alder district.
Alder_Districts['coords'] = Alder_Districts['geometry'].apply(lambda x: x.representative_point().coords[:])
Alder_Districts['coords'] = [coords[0] for coords in Alder_Districts['coords']]

bp = mc.UserDefined(Alder_Districts["Density"], bins=[20, 30, 40, 50, 60, 80, 90, 100])
ax = Alder_Districts.plot(figsize = (20, 15), legend=True, scheme = 'UserDefined',
                  classification_kwds={'bins': [20, 30, 40, 50, 60, 80, 90, 100]},
                  k = 5, column='Density', cmap = 'Blues')

for idx, row in Alder_Districts.iterrows():
    ax.text(row.coords[0], row.coords[1], s=int(row['ALD_DIST']), horizontalalignment='center', bbox={'facecolor': 'white', 'alpha':0.8, 'pad': 2, 'edgecolor':'none'})

Alder_Districts.boundary.plot(ax=ax, cmap='rainbow_r')
handles, labels = ax.get_legend_handles_labels()
cmap = plt.get_cmap('Blues')
bins = bp.bins
bins = np.insert(bins, 0, 0)
LegendElement = [mpatches.Patch(facecolor=cmap(_*0.25), label=f'{max(bins[_], 0)}% - {bins[_+1]}%')
                 for _ in range(7)] + [mpatches.Patch(facecolor='#00008b', hatch='', label='90%+')]
ax.legend(handles = LegendElement, loc='upper left', fontsize=25, shadow=True, borderpad=0.3, bbox_to_anchor= (0.1, 0.95))
ax.set_axis_off()
ax.set_title('Percentage of having different tax address in Madison area', fontsize = 30)
```
