```python
import pandas as pd
import re
```

```python
table = pd.read_csv("table.csv")
```

```python
table
```

```python
newAddress = []
```

```python
for i in table['Address']:
    if bool(re.search('-\d\d\d\d$', i)):
        newAddress.append(i[:-5])
    else:
        newAddress.append(i)
```

```python
table['newAddress'] = newAddress
```

```python
table
```

```python
table.sort_values("newAddress", inplace = True)
bool_series = table.duplicated(subset = ['newAddress'], keep = False)
```

```python
sum(bool_series)
```

```python
table[bool_series]
```

```python
table[bool_series]['newAddress']
```

```python
dic = {}
for i in table[bool_series]['newAddress']:
    if i not in dic:
        dic[i] = 0
    dic[i] += 1
```

```python
dic
```

```python
dic = dict(sorted(dic.items(), key=lambda item: item[1], reverse=True))
```

```python
cumulativeList = []
count = 0
for i in dic:
    count += dic[i]
    cumulativeList.append(count)
```

```python
countList = []
for count, i in enumerate(cumulativeList):
    countList.append(count+1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(countList)
ypoints = np.array(cumulativeList)

plt.plot(xpoints, ypoints)
plt.show()
```
