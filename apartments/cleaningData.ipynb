```python
import pandas as pd
```

```python
table = pd.read_csv("Tax_Parcels.csv")
table = table.fillna(0)
```

```python
table
```

```python
residentialTable = table[table['PropertyClass'] == 'Residential']
```

```python
residentialTable
```

```python
commercialTable = table[table['PropertyClass'] == 'Commercial']
commercialTable
```

```python
def revise(use):
    use = str(use)
    if len(use) > 14 and use[-14:] == 'unit Apartment':
        return 'unit apartment'
    elif use == 'Commercial exempt':
        return 'commercial exempt'
    elif use[:11].lower() == 'condominium':
        return 'condominium'
    else:
        return 'else'
```

```python
x = commercialTable['PropertyUse'].apply(revise)
commercialTable.insert(0, "PropertyUseOrElse", x)
```

```python
commercialTable[commercialTable['PropertyUseOrElse'] == 'unit apartment']
commercialTable[commercialTable['PropertyUseOrElse'] == 'commercial exempt']
commercialTable[commercialTable['PropertyUseOrElse'] == 'condominium']
```

```python
result = pd.concat([residentialTable, 
                    commercialTable[commercialTable['PropertyUseOrElse'] == 'unit apartment'],
                    commercialTable[commercialTable['PropertyUseOrElse'] == 'commercial exempt'],
                    commercialTable[commercialTable['PropertyUseOrElse'] == 'condominium']])
result = result[['OBJECTID', 'Parcel', 'Address', 'PropertyClass', 'PropertyUse', ]]
```

```python
result.to_csv('cleandata.csv', index=False)
```
