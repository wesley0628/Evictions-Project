```python
import pandas as pd
```

```python
cases = pd.read_csv("data/ccapoutputfile_20210706023712.csv")
cases
```

```python
landlord = cases.groupby(["Name"]).count().iloc[:, :1]
landlord = landlord.rename(columns={"CaseNumber": "Count"})
landlord = landlord.sort_values(by=['Count'], ascending=False)
landlord[:20]
```

```python
zipcode = cases.groupby(["ZipCodes"]).count().iloc[:, :1]
zipcode = zipcode.rename(columns={"CaseNumber": "Count"})
zipcode = zipcode.sort_values(by=['Count'], ascending=False)
zipcode[:20]
```
