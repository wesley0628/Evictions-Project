```python
def scrape(columns):
    index = 0
    plaintiffList=[]
    while index < len(columns.text):
            index = columns.text.find('Plaintiff', index)
            if index == -1:
                break
            plaintiffList.append(index)
            index += 2
    index = 0
    defendantList = []
    while index < len(columns.text):
            index = columns.text.find('Defendant', index)
            if index == -1:
                break
            defendantList.append(index)
            index += 2

    plaintiffName = []
    defendantName = []
    if plaintiffList[0]<defendantList[0]:
        for i in range(len(plaintiffList)):
            if i == len(plaintiffList) - 1:
                plaintiffName.append(columns.text[plaintiffList[i]+9:defendantList[0]].strip())
                break
            plaintiffName.append(columns.text[plaintiffList[i]+9:plaintiffList[i+1]].strip())

        for i in range(len(defendantList)):
            if i == len(defendantList) - 1:
                defendantName.append(columns.text[defendantList[i]+9:].strip())
                break
            defendantName.append(columns.text[defendantList[i]+9:defendantList[i+1]].strip())
        return plaintiffName, defendantName
    else:
        for i in range(len(defendantList)):
            if i == len(defendantList) - 1:
                defendantName.append(columns.text[defendantList[i]+9:plaintiffList[0]].strip())
                break
            defendantName.append(columns.text[defendantList[i]+9:defendantList[i+1]].strip())

        for i in range(len(plaintiffList)):
            if i == len(plaintiffList) - 1:
                plaintiffName.append(columns.text[plaintiffList[i]+9:].strip())
                break
            plaintiffName.append(columns.text[plaintiffList[i]+9:plaintiffList[i+1]].strip())
        return plaintiffName, defendantName
```

```python
import time
casesList = []
plaintiffsList = []
defendantsList = []
errorCount = 0
errorList = []
count = 1
for i in table['CaseNumber']:
    print(count)
    count += 1
    try:
        driver.get('https://wcca.wicourts.gov/caseDetail.html?caseNo={}&countyNo=13'.format(i))
        time.sleep(5)
        with open('cases\{}.html'.format(i), 'w') as f:
            f.write(driver.page_source)
        columns = driver.find_element_by_class_name('columns')
        plaintiffs, defendants = scrape(columns)
        casesList.append(i)
        plaintiffsList.append('; '.join(plaintiffs))
        defendantsList.append('; '.join(defendants))
    except:
        errorCount += 1
        errorList.append(i)

data = {"Caes": pd.Series(casesList),
        "plaintiffs": pd.Series(plaintiffsList),
        "defendants": pd.Series(defendantsList)}

df = pd.concat(data, axis = 1)
df.to_csv('table.csv', index = False)
print('errorCount:', errorCount)
print('errorList:', errorList)
```
