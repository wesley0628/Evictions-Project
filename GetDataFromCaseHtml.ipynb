```python
import requests
import pandas as pd
import html5lib
from lxml import html
from bs4 import BeautifulSoup
```

```python
file = open("case_html/" + str(2) + "case.csv", "r")
content = file.read()
soup = BeautifulSoup(content, 'html5lib')
details = soup.find(id="civilJdgmts")
caseNo = details.find("a", {"class": "case-link"})
caseNo.text
```

```python
# find Responsible official and judgment status
def get_judge_result(df):
    
    details = soup.find(id="civilJdgmts")
    if details is None:
        #return df
        caseNo = 0
    else:
        caseNo = details.find("a", {"class": "case-link"}).text
       
    summary = soup.find(id='summary')
    if summary is None:
        return df
    judge = summary.findAll('dd')[4].text
    rows = summary.findAll('td')
    result = False
    for row in rows:
        text = row.text
        if "judg" in text:
            df.loc[len(df.index)] = [judge, text, caseNo] 
            result = True
        elif "satis" in text:
            df.loc[len(df.index)] = [judge, text, caseNo] 
            result = True
    if result == False:
        df.loc[len(df.index)] = [judge, 'NA', caseNo] 
    return df
```

```python
small_df = pd.DataFrame()
small_df["judge"] = []
small_df["result"] = []
small_df["caseNo"] = []

for i in range(1, 85):
    file = open("case_html/" + str(i) + "case.csv", "r")
    content = file.read()
    soup = BeautifulSoup(content, 'html5lib')
    small_df = get_judge_result(small_df)

small_df
```
